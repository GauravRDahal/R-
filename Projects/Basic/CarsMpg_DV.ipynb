Initating the ggplot() binding to the car dataframe

```r
pl = ggplot(data = mpgData)
pl
```

```r
# Using the aes() to specify the sedthetic mapping
pl = ggplot(data = mpgData, aes(x = displ, y = hwy))
```

```r
# Using geom_ type function to add geometric elements
pl = ggplot(data = mpgData, aes(x = displ, y = hwy)) + geom_point()
```

```r
# Adding labels and title
pl = pl + labs(x = "Displacement (liters)", y = "Mileage (MPG)", title = "Mileage vs. Displacement")
pl
```

Mapping aesthetics to variables

```r
# mapping the color asthetics to the class variable
p2 = ggplot(data = mpgData) + geom_point(mapping = aes(x = displ, y = hwy, color = class))
p2
```

```r
# mapping the size (or alpha or shape) aesthetics to the class variable
pl3 = ggplot(data = mpgData) + geom_point(mapping = aes(x = displ, y = hwy, size = class))
pl3
```

```r
pl4 = ggplot(data = mpgData) + geom_point(mapping = aes(x = displ, y = hwy, alpha = class), size = 5)
pl4
```

```r
pl5 = ggplot(data = mpgData) + geom_point(mapping = aes(x = displ, y = hwy), color = 'blue', size = 2)
pl5
```

Adding additional variables using facets.

```r
# plotting mileage w.r.t. each class individual
ggplot(data = mpgData) + geom_point(mapping = aes(x = displ, y = hwy)) + facet_wrap(~class, nrow = 4)
```

```r
# plotting mileage w.r.t. drive train and number of cylinders
ggplot(data = mpgData) + geom_point(mapping = aes(x = displ, y = hw)) + facet_grid(drv ~ cyl)
```

```r
# Investigating the number of levels in the cyl feature
factor(mpgData$cyl)
```

```r
# How many cars of each cyl type are there?
mpgData %>% count(cyl)
```

```r
# Filtering samples with 5 cylinder cars (Which are rare)
mpgData = mpgData %>% filter(!cyl %in% c(5))
# this means to filter only cylinder value that's equal to 5
head(mpgData, 5)
```

```r
# mapping the color aesthetic to the cyl variable
ggplot(data = mpgData) + geom_point(mapping = aes(x = displ, y = hwy, color = factor(cyl)))
```

```r
ggplot(data = mpgData) + geom_point(mapping = aes(x = displ, y = hwy, color = factor(cyl), shape = class))
```

```r
# plotting a smooth line plot of mileage w.r.t. displacement
ggplot(data = mpgData) + geom_smooth(mapping = aes(x = displ, y = hwy))
```

```r
# shaded area is standard error
ggplot(data = mpgData) + geom_smooth(mapping = aes(x = displ, y = hwy), se = FALSE)
```

```r
ggplot(data = mpgData) + geom_smooth(mapping = aes(x = displ, y = hwy, linetype = drv), se = FALSE)
```

```r
#with respect to cylinder
ggplot(data = mpgData) + geom_smooth(mapping = aes(x = displ, y = hwy, linetype = factor(cyl)), se = FALSE)
```

Specifying data for layers individually

```r
#plotting mileage w.r.t. displacement for all cars but adding a smooth line only for subcompact cars by filtering
 ggplot(data = mpgData, mapping = aes(x = displ, y = hwy)) +
   geom_point(mapping = aes(color = factor(cyl))) + geom_smooth(data = filter(mpgData, class == 'subcompact'), se = FALSE)
```

```r
# adding a bar chart w.r.t. number of cylinders
ggplot(data = mpgData) + geom_bar(mapping = aes(x = cyl))
```

```r
# adding a boxplot w.r.t number of cylinders and mileage
ggplot(data = mpgData) + geom_boxplot(mapping = aes(x = factor(cyl), y = hwy))
```

```r
# changing outlier color, shape and size
ggplot(data = mpgData) + geom_boxplot(mapping = aes(x = factor(cyl), y = hwy), outlier.color = 'red', outlier.shape = 8, outlier.size = 4)
```

Plotting a basic histogram

```r
df = data.frame(
  gender = factor(rep(c("F","M"), each = 200)),
  height = round(c(rnorm(200, mean = 60,sd = 10), rnorm(200, mean = 70, sd = 6)))
)
```

```r
#head(df,5)
str(df)
```

```r
ggplot(data = df, mapping = aes(x = height)) + geom_histogram()
```

```r
# changing the width of bins
ggplot(data = df, mapping = aes(x = height)) + geom_histogram(binwidth = 5)
```

```r
# change the color
ggplot(data = df, mapping = aes(x = height)) + geom_histogram(color = 'red', fill = 'blue')
```

```r
# changing histogram plot line colors by gender
ggplot(df, aes(x = height, color = gender)) + geom_histogram(fill = 'grey')
```

```r
#overlaying histograms for both genders
ggplot(data = df, mapping = aes(x = height, color = gender)) +
  geom_histogram(fill = 'white', alpha = 0.8, position = 'identity')
```

```r
# plotting a bar chart of proportion, rather than count w.r.t. number of cylinders
ggplot(mpgData) +
  geom_bar(aes(x = cyl, y = ..prop..))
```

```r
# plotting a histogram of engine displacement
ggplot(data = mpgData) + geom_histogram(mapping = aes(x =displ, y = ..count..),
binwidth = 0.5, color = 'black', fill = 'steelblue', alpha = 0.4)
```

```r
# plotting a histogram of engine displacement
ggplot(data = mpgData) + geom_histogram(mapping = aes(x =displ, y = ..density..),
binwidth = 0.5, color = 'black', fill = 'steelblue', alpha = 0.4)
```

```r
# plotting a bar chart w.r.t. number of cylinders and class using color and fill
ggplot(data = mpgData) + geom_bar(aes(x = cyl, fill = class))
```

```r
ggplot(data = mpgData) + geom_bar(aes(x = cyl, fill = class), position = 'fill')
```
