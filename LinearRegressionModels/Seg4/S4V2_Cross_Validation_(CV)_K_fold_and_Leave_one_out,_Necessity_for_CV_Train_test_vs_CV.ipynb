# Cross Validation (CV)

Cross validation is a technique for estimating prediction error of a model when the amount of data vailable is small.

- Basic idea is to split the data sirst into training and test parts.
- followed by subsetting the `training` part into `subtrain` and `validation` parts
- and cycling through those subsets in a round-robin fashion for mdoel building

- Each block will provide us with a single prediction error.

## **K-fold Cross Validation**

Here, k is an ineteger that represents the splits (number of subsets).

- The number of subsets `k` into which the training data is split results in different types of cross validation approaches.
- Each subset of the training data is referred to as a `fold`.
- When `K` is much smaller than the number of training samples, the model are built using `subtrain`-folds with a small number of samples.

The effect is that because a small number of samples are used for building the model, the prediction error that will be obtained using the model that we will build will be an over estimate of the true prediction error of the eventual model that we will have built using the entire training dataset.(High Bias)

## **Leave-one-out Cross Validation**

When `k=n`, the resulting cross validation approach is called the **leave-one-out cross validation** procedure.

- The advantage of this procedure is that we make use of almost the entire `training` data for model building which results in an almost unbiased estimate of the true prediction error. (**Low Bias**)

- on the other hand, the `leave-one-out` cross validation procedure will result in an estimate of the true prediction error that is very sensitive to the `subtrain` data used to train the models.

The statistical reason is that all the models will essentially be looking at almost the same training data. or, the prediction error that comes out of all the models that we build inside the cross validation preocedure will be highly correlated/ positively correlated. And when we average positively correlated things, the variance tends to be higher. Therefore, true prediction error is very sensitive to the `subtrain` data used to train the models (**High Variance**)

### Due to this reason, `k` is typically chosen to be `5` or `10` depending on the complexity of the model to strike a balance between computational efficiency and a reliable estimate of the true prediction error.

Another reseon for choosing the `leave-one-out` cross validation procedure in the linear regression scenario is that we don't have to rebuild the models in order to calculate the cross-validation procedures prediction error.

# **Train-test-Validation split vs. CV**

How do we decide to choose between a train-validation-test preocedure and a cross-validation procedurr?

In production systems, the model will be trained on the entire data that available.
- In this scenario, if the computational cost is not too high, it would be better to use a cross validation procedure as it would yield a reliable (low-variance) estimate of the generalization error.
- On the other hand, if the data has a seperate substantial amount of data as the test set that we do not have access to, we can use the train-validation split which will lead to an unbiased estimate of the generalization error.
